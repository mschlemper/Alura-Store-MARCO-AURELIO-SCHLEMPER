## Transforma√ß√£o dos dados

### Subtask:
Achatar as colunas aninhadas ('customer', 'phone', 'internet', 'account') no DataFrame principal para facilitar a an√°lise.

**Reasoning**:
The columns 'customer', 'phone', 'internet', and 'account' contain nested dictionary-like structures. These need to be flattened into individual columns to make the data accessible for analysis.

```python
# Flatten the 'customer' column
customer_df = pd.json_normalize(df_telecom['customer'])
df_telecom = pd.concat([df_telecom.drop('customer', axis=1), customer_df], axis=1)

# Flatten the 'phone' column
phone_df = pd.json_normalize(df_telecom['phone'])
df_telecom = pd.concat([df_telecom.drop('phone', axis=1), phone_df], axis=1)

# Flatten the 'internet' column
internet_df = pd.json_normalize(df_telecom['internet'])
df_telecom = pd.concat([df_telecom.drop('internet', axis=1), internet_df], axis=1)

# Flatten the 'account' column
account_df = pd.json_normalize(df_telecom['account'])
df_telecom = pd.concat([df_telecom.drop('account', axis=1), account_df], axis=1)

# Display the first few rows of the transformed DataFrame to verify
print("DataFrame ap√≥s 'achatar' as colunas aninhadas:")
display(df_telecom.head())

# Display information about the transformed DataFrame
print("\nInforma√ß√µes do DataFrame transformado:")
df_telecom.info()
```

### Subtask:
Converter a coluna 'Charges.Total' para um tipo num√©rico, tratando poss√≠veis valores n√£o num√©ricos que impediram a convers√£o autom√°tica durante o carregamento.

**Reasoning**:

Non-numeric values (potentially empty strings) should be handled, likely by converting them to NaN and then filling with 0, as per common practice with such columns in telecom datasets where empty total charges might indicate new customers.

```python
# Convert 'Charges.Total' to numeric, coercing errors and filling NaNs with 0
df_telecom['Charges.Total'] = pd.to_numeric(df_telecom['Charges.Total'], errors='coerce').fillna(0)

# Verify the data type after conversion
print("Tipo de dados da coluna 'Charges.Total' ap√≥s a convers√£o:")
print(df_telecom['Charges.Total'].dtype)

# Display descriptive statistics for 'Charges.Total' to confirm conversion
print("\nEstat√≠sticas descritivas para a coluna 'Charges.Total':")
display(df_telecom['Charges.Total'].describe())
```

### Subtask:
Realizar an√°lises explorat√≥rias b√°sicas e visualiza√ß√µes para entender as caracter√≠sticas dos dados e identificar padr√µes.

**Reasoning**:
With the data loaded, flattened, and cleaned (handling missing values and data types), perform some basic exploratory data analysis (EDA) to gain insights into the dataset.
