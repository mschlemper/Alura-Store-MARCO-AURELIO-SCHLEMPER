**Reasoning**:
I need to import the `PdfReader` class from `PyPDF2` and then use it to open the PDF file, iterate through its pages, and extract all the text. The extracted text will be stored in a variable for further analysis.



````python
from PyPDF2 import PdfReader

pdf_file_path = '/content/Desenvolvimento de IA para Análise Preditiva [T2] - PROJETO FINAL MARCO A SCHLEMPER.pdf'
pdf_text = ''

try:
    with open(pdf_file_path, 'rb') as file:
        reader = PdfReader(file)
        for page_num in range(len(reader.pages)):
            page = reader.pages[page_num]
            pdf_text += page.extract_text() + "\n"
    print(f"Successfully extracted text from {len(reader.pages)} pages.")
    print("First 2000 characters of the extracted text:\n")
    print(pdf_text[:2000])
except Exception as e:
    print(f"An error occurred: {e}")
````
